```python
def isTalk(a_string):
    p1 = len(re.findall(r"^\w+\t", a_string))
    p2 = len(re.findall(r"^\w+\s+[()]", a_string))
    res = p1 + p2
    if res > 0:
        return True
    else:
        return False
```

```python
df['isTalk'] = df['script'].map(isTalk)
```
